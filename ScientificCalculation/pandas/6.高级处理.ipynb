# 缺失值处理

```python
movie = pd.read_csv("./data/IMDB-Movie-Data.csv")
```

```python
movie
```

```python
np.any(pd.isnull(movie))  # 里面如果有一个缺失值,就返回True
```

```python
np.all(pd.notnull(movie))  # 里面如果有一个缺失值,就返回False
```

```python
data = movie.dropna()
```

```python
np.any(pd.isnull(data))  # 里面如果有一个缺失值,就返回True
```

```python
movie["Revenue (Millions)"].head(10)
```

```python
# movie["Revenue (Millions)"].fillna(value=movie["Revenue (Millions)"].mean(), inplace=True)
```

```python
for i in movie.columns:
    # print(i)
    if np.any(pd.isnull(movie[i])) == True:
        # print(i)
        movie[i].fillna(value=movie[i].mean(), inplace=True)
```

```python
np.any(pd.isnull(movie))  # 里面如果有一个缺失值,就返回True
```

# 数据离散化处理

```python
data = pd.read_csv("./data/stock_day.csv")
```

```python
data_p = data["p_change"]
```

```python
data_p.head()
```

```python
# help(pd.qcut)
```

```python
qcut_r = pd.qcut(data_p, q=10)
```

```python
qcut_r.value_counts()
```

```python
bins = [-100, -7, -5, -3, 0, 3, 5, 7, 100]
cut_r = pd.cut(data_p, bins=bins)
```

```python
ret = cut_r.value_counts()
```

```python
ret
```

```python
cut_r.head()
```

```python
data_dummies = pd.get_dummies(cut_r, prefix="python38_best")
```

# 交叉表和透视表

```python
data.index
```

```python
time = pd.to_datetime(data.index)
```

```python
time.day
```

```python
time.week
```

```python
data["week"] = time.weekday
```

```python
data.head()
```

```python
data["p_n"] = np.where(data["p_change"]>0, 1, 0)
```

```python
data.head()
```

```python
count = pd.crosstab(data["week"], data["p_n"])
```

```python
count
```

```python
sum = count.sum(axis=1)
```

```python
sum
```

```python
per = count.div(sum, axis=0)
```

```python
per
```

```python
per.plot(kind="bar", stacked=True)
```

```python
data.pivot_table(["p_n"], index="week")
```

# 分组和聚合

```python
col =pd.DataFrame({'color': ['white','red','green','red','green'], 'object': ['pen','pencil','pencil','ashtray','pen'],'price1':[5.56,4.20,1.30,0.56,2.75],'price2':[4.75,4.12,1.60,0.75,3.15]})
```

```python
col
```

```python
col.groupby(["color"])["price1"].mean()
```
